### Use json to get boot images

```python
import urllib
import os

import coremltools 
from skafossdk import *
import turicreate as tc
import pandas as pd
from s3fs.core import S3FileSystem
```

```python
s3 = S3FileSystem(anon=True)
json_path = "s3://mf.testing.bucket/Boots/F21_shoes.json"

df = pd.read_json(json_path)
```

```python
components = pd.DataFrame.from_dict(df.iloc[8].body)
components_df = pd.DataFrame.from_records(components.docs)
```

```python
components_df['path'] = components_df['thumb_image'].apply(lambda x: 'BootImages/' + x.split("/")[-2:][1])
```

```python
components_df.head()
```

```python
for index, row in components_df.iterrows():
    urllib.request.urlretrieve(row['thumb_image'], row['path'])
```

```python
F21_meta = tc.SFrame(components_df)
```

```python
F21_meta = F21_meta.rename({'title': 'Description'})
F21_meta['Image Name'] = F21_meta['path'].apply(lambda x: x.split("/")[-2:][1])
```
